# Movement

```python
movement_data = pd.read_csv('./movement/test_data/test2.csv', header=None)
```

```python
plt.plot(movement_data[0].iloc[:])
```

```python
move_tracker = []
for ind in range(0,len(movement_data[0])):
    if movement_data[0].iloc[ind]>73:
        move_tracker.append(70)
    else:
        move_tracker.append(0)
```

```python
movement_data[1] = move_tracker
movement_data.head()
```

```python
plt.plot(movement_data[0].iloc[:])
plt.plot(movement_data[1].iloc[:])
plt.show()
```

```python
mark = 500
ind = 0
first_step = np.zeros((1, len(movement_data[0])))
while ind < len(movement_data[0]):
    if np.sum(movement_data[1].iloc[ind:ind+mark])>=9000: #6600
        first_step[0,ind:ind + mark] = 70
    ind = ind + 50
```

```python
movement_data[2] = list(first_step[0,:])
movement_data.head()
```
